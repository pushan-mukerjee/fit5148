```python
# Libraries
import socket
import sys
import time
import random
import datetime as dt
import pandas as pd
import json
```

```python
#Reading the ClimateData-Part2.csv into an array of values
climateDf=pd.read_csv('ClimateData-Part2.csv', sep=',',header=0)
climateRecords = climateDf.values
climateRecords = climateRecords.tolist()

#Reading the FireData-Part2.csv into an array of values
fireDf=pd.read_csv('FireData-Part2.csv', sep=',',header=0)
fireRecords = fireDf.values
fireRecords = fireRecords.tolist()
```

```python

# Create a TCP/IP socket
#climateSock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
fireSock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Bind the socket to the port
#server_address = ('localhost', 8085)
#print('['+ str(dt.datetime.now())+'] Starting streaming server on {}:{}'.format(*server_address))
#climateSock.bind(server_address)

# Bind the socket to the port
server_address2 = ('localhost', 9997)
print('['+ str(dt.datetime.now())+'] Starting streaming server on {}:{}'.format(*server_address2))
fireSock.bind(server_address2)

# Listen for incoming connections
#climateSock.listen(1)
fireSock.listen(1)

i = 0
while True:
    # Wait for a connection
    print('['+ str(dt.datetime.now())+'] Waiting for a connection...')
    #connection, client_address = climateSock.accept()
    connection2, client_address2 = fireSock.accept()
    try:
        #print('['+ str(dt.datetime.now())+'] Connection from', client_address)
        print('['+ str(dt.datetime.now())+'] Connection from', client_address2)
        # Get the random line from lines array
        datestamp = str(dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
        #climateRec = climateRecords[random.randrange(366)]

        #if (i < len(climateRecords)):
        #    climateRec = climateRecords[i] #pick the next climateData record to send
        #    climateRec.append(datestamp)
        #    connection.sendall(str(climateRec).encode())
        #    print ('Climate Data Sent: ' + str(climateRec))
        
        if (i < len(fireRecords) // 5):
            startRec = 5 * i #define the start record index of the fireData batch
            if (len(fireRecords) - startRec < 5): #if the last batch of fireData records, get all remaining records
                endRec = (5*i) + (len(fireRecords) - startRec)
            else: #Else pick the next 5 fireData records to send in the fire batch
                endRec = (5*i) + 5
            fireRecs = fireRecords[startRec:endRec]

            for fireRec in fireRecs: #Loop through the fireData records and add them to the batch of 5 to be sent
                fireRec.append(datestamp)
                connection2.sendall(str(fireRec).encode()) #Send the FireData batch
                print ('FireData Sent: ' + str(fireRec)) #Print the FireData batch
        i += 1    
        time.sleep(1)
    finally:
        # Clean up the connection
        #connection.close()
        connection2.close()
```
